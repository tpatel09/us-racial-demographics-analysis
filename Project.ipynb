```python
import matplotlib.pyplot as plt 

# Define a threshold for the 'White Alone' population
threshold = 0.4

# Filter out districts where the 'White Alone' population exceeds the threshold
filtered_df = clean_df[clean_df['White Alone'] / clean_df['Total Population'] <= threshold]

for state_district, group in filtered_df.groupby(['State', 'District Name']):
    title = str(state_district)
    group.plot(x='Year', y=['White Alone', 'Black Alone', 'American Indian or Alaskan Native', 'Asian Alone', 'Hawaiian or Pacific Islander Alone'], kind='line', title=title)
    plt.show()
```

```python
district_df = clean_df[clean_df['District Name'] == 'Northwest New Mexico Council of Governments']

# Select the rows for the years 1990 and 2018
district_1990_df = district_df[district_df['Year'] == 1990]
district_2018_df = district_df[district_df['Year'] == 2018]

# Select the columns representing the racial groups
racial_cols = ['White Proportion', 'Black Proportion', 'American Indian or Alaskan Native Proportion', 'Asian Proportion', 'Hawaiian or Pacific Islander Alone Proportion', 'Two or More Races Proportion', 'Not Hispanic Proportion', 'Hispanic Proportion']

# Create the bar plots
district_1990_df[racial_cols].plot(kind='bar', title='Racial Composition of Northwest New Mexico Council of Governments in 1990')
plt.legend(fontsize='small') 
plt.show()

district_2018_df[racial_cols].plot(kind='bar', title='Racial Composition of Northwest New Mexico Council of Governments in 2018')
plt.legend(fontsize='small') 
plt.show()
```

```python
# Define the two years to compare
year1 = 2000
year2 = 2019

for state_district, group in filtered_df.groupby(['State', 'District Name']):
    title = str(state_district)
    
    # Filter the data for the specific years
    district_year1_df = group[group['Year'] == year1]
    district_year2_df = group[group['Year'] == year2]
    
    # Plot racial composition for the specific district and years
    if not district_year1_df.empty and not district_year2_df.empty:
        fig, axs = plt.subplots(2)
        fig.suptitle(title)
        district_year1_df[['White Proportion', 'Black Proportion', 'American Indian or Alaskan Native Proportion', 
                           'Asian Proportion', 'Hawaiian or Pacific Islander Alone Proportion']].plot(kind='bar', 
                            ax=axs[0], title=f'Year {year1}')
        district_year2_df[['White Proportion', 'Black Proportion', 'American Indian or Alaskan Native Proportion', 
                           'Asian Proportion', 'Hawaiian or Pacific Islander Alone Proportion']].plot(kind='bar', 
                            ax=axs[1], title=f'Year {year2}')
        
        # Make the legend smaller
        axs[0].legend(fontsize='small')
        axs[1].legend(fontsize='small')
        
        plt.show()
```
